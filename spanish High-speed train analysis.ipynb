```python
import pandas as pd
train_data = pd.read_csv("renfe.csv")
train_data.head()
```

```python
train_data.shape
```

### 数据去空

```python
train_data.isna().any()
```

```python
train_data = train_data.dropna()
train_data.shape
```

```python
train_data["day"] = train_data["start_date"].str[5:10]
train_data["hour"] = train_data["start_date"].str[11:]
```

### 出行人数最多是哪一天

```python
checi = train_data.groupby("day")["insert_date"].count()
checi.describe()
```

```python
checi.idxmax()
```

```python
checi
```

### 大家去了哪里

```python
train_data.loc[train_data["day"]=="05-30","destination"].value_counts()
```

### 去掉人数 统计班车信息

```python
train_price_wave = train_data.drop("insert_date",axis =1)
```

```python
train_price_wave.shape
```

```python
train_price_wave = train_price_wave.drop_duplicates()
train_price_wave.shape
```

### 价格波动分析

```python
train_price_wave["start_date"].max()
```

```python
train_price_wave["start_date"].min()
```

```python
train_price_wave_2019 = train_price_wave.loc[(train_price_wave["start_date"]>="2019-04-00")&(train_price_wave["start_date"]<="2019-10-30")]
train_price_wave_2019.shape
```

### 马德里到巴塞罗那的票价波动

```python
M_to_B = train_price_wave_2019.loc[(train_price_wave_2019["origin"]=="MADRID")&(train_price_wave_2019["destination"]=="BARCELONA")]
M_to_B.shape
```

```python
M_to_B
```

```python
M_to_B["train_class"].value_counts()
```

### 我们发现 同一班车每天不同时刻发车的票价 也是不同的。那我们对当天的票价进行一个平均处理。

```python
M_to_B_Turista = M_to_B.loc[M_to_B["train_class"]=="Turista"]
M_to_B_Turista_price = M_to_B_Turista.groupby("day")["price"].mean().round(0)
M_to_B_Turista_price
```

### 如果想知道哪天哪个时刻票最便宜

```python
M_to_B_Turista_price_min = M_to_B_Turista.groupby("day")["price"].min().round(0)
M_to_B_Turista_price_min =M_to_B_Turista_price_min[M_to_B_Turista_price_min>0]
M_to_B_Turista_price_min
```

```python
x,y = M_to_B_Turista_price_min.index.tolist(),M_to_B_Turista_price_min.values.tolist()
line = (
    Line()
    .add_xaxis(x)
    .add_yaxis("票价", y,label_opts=opts.LabelOpts(is_show=False),
               markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(type_="min")]),
              )

    .set_global_opts(title_opts=opts.TitleOpts(title=""), xaxis_opts=opts.AxisOpts(
            axistick_opts=opts.AxisTickOpts(is_align_with_label=False),
            axislabel_opts={"rotate":-90},
            is_scale=False,
            boundary_gap=False,
        ),)
)
line.render("全年最低票价.html")
```

## 全年价格波动

```python

from pyecharts import options as opts
from pyecharts.charts import Bar, Grid, Line,Scatter

from pyecharts import options as opts
from pyecharts.charts import Bar

x,y,z = checi.index.tolist(),checi.values.tolist(),M_to_B_Turista_price.values.tolist()

bar = (
    Bar()
    .add_xaxis(x)
    .add_yaxis("出行人数",y,  markpoint_opts=opts.MarkPointOpts(
                data=[opts.MarkPointItem(type_="max")]
            ),
              )                                                       
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    
    .set_global_opts(
        title_opts=opts.TitleOpts(title="出行人数"),
        xaxis_opts=opts.AxisOpts(
            axistick_opts=opts.AxisTickOpts(is_align_with_label=False),
            axislabel_opts={"rotate":-90},
            is_scale=False,
            boundary_gap=False,
        ),
        
    )
    .extend_axis(
        yaxis=opts.AxisOpts(
            name="票价",
            type_="value",
            min_=0,
           # max_=1.00,
            position="right",
            axisline_opts=opts.AxisLineOpts(
                linestyle_opts=opts.LineStyleOpts(color="#d14a61")
            ),
            axislabel_opts=opts.LabelOpts(formatter="{value}"),
        )
    )
 )
line = (
    Line()
    .add_xaxis(x)
    .add_yaxis("票价", z,yaxis_index=1,label_opts=opts.LabelOpts(is_show=False),
               markpoint_opts=opts.MarkPointOpts(
                   data=[opts.MarkPointItem(type_="max")]
               ),
              )

    .set_global_opts(title_opts=opts.TitleOpts(title=""),)
)
   


bar.overlap(line)
grid = (
        Grid()
        .add(bar, opts.GridOpts(pos_left="10%", pos_right="10%"), is_control_axis_index=True)
        
    )


grid.render("出行人数和票价波动.html")

```
